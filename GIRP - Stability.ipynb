```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
portugese_results = pd.read_excel('stability results.xlsx')
print(portugese_results.columns)

german_results = pd.read_excel('stability results.xlsx', sheet_name = 'German Credit')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(portugese_results['Kappa'], portugese_results['Top-j'], color=(0.1, 0.2, 0.5, 0.3), s = (portugese_results['Number of Nodes']*5))
ax.scatter(german_results['Kappa'], german_results['Top-j'], color=(0.0, 0.48, 0.52, 0.3), s = (german_results['Number of Nodes']*5))

ax.set_xlabel('Cohen\'s Kappa')
ax.set_ylabel('Top-j Similarity')
ax.set_title('Stability GIRP')
#ax.legend('Portugese', 'German')
plt.show()
```
